# DecisionTreeRegressor

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

df = pd.read_csv('coronavirus-data-explorer.csv', index_col = 'date',parse_dates = True)

df.head()
```

```python
uk = df[df['iso_code']== 'GBR']
```

```python
#clean data ie remove NaN's

uk_clean = uk.replace(np.nan,0)
```

```python
uk_clean.head()
```

```python
#Features used to train data

features = [ 'new_cases', 'new_deaths', 'reproduction_rate', 'hosp_patients', 'new_tests', 'tests_per_case', 'positive_rate','people_vaccinated','people_fully_vaccinated','new_vaccinations','stringency_index']
```

```python
#The target variable

output = ['icu_patients']
```

```python
x = uk_clean[features]
```

```python
y = uk_clean[output]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
```

```python
regressor.fit(x,y)
```

```python
y_pred = regressor.predict(x_test)
```

```python
#regression score

from sklearn.model_selection import cross_val_score
score = regressor.score(x_test, y_test)
print (score)   
```

```python
#cross-validation

score = cross_val_score(regressor, x_test, y_test, cv =10)
print (score)  
```

```python
#root-mse error

from sklearn import metrics

print('Root Mean Squared Error is:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
#ranking of importance features

prediction_importance = list(regressor.feature_importances_)
input_importance = [(feature, importance) for feature, importance in zip(features, prediction_importance)]
[print('For feature {} Relative Importance is {}'.format(*rows)) for rows in sorted(input_importance)];
```

# Random Forest

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split


df = pd.read_csv('coronavirus-data-explorer.csv', index_col = 'date',parse_dates = True)

df.head()

```

```python
#get UK data and clean errors

uk = df[df['iso_code']== 'GBR']

uk_clean = uk.replace(np.nan,0)

uk_clean.head()
```

```python
#features used

features = [ 'new_cases', 'new_deaths', 'reproduction_rate', 'hosp_patients', 'new_tests', 'tests_per_case', 'positive_rate','people_vaccinated','people_fully_vaccinated','new_vaccinations','stringency_index']
```

```python
#target variable

output = ['icu_patients']
```

```python
x = uk_clean[features]
```

```python
y = uk_clean[output]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor

# Create the model with n trees

RFC = RandomForestRegressor(n_estimators=100, bootstrap = True,max_features = 'auto' )
```

```python
RFC.fit(x, y.values.ravel())

```

```python
y_pred = RFC.predict(x_test)
```

```python
#regression score

from sklearn.model_selection import cross_val_score
score = RFC.score(x_test, y_test.values.ravel())
#score = RFC.score(x_test, y_test)
print (score)
```

```python
#cross-validation score

score = cross_val_score(RFC, x_test, y_test.values.ravel(), cv =10)
print (score) 
```

```python
#root mean square error

from sklearn import metrics

print('Root Mean Squared Error is:', np.sqrt(metrics.mean_squared_error(y_test.values.ravel(), y_pred)))
```

```python
#ranking of feature importance

prediction_importance = list(RFC.feature_importances_)
input_importance = [(feature, importance) for feature, importance in zip(features, prediction_importance)]
[print('For feature {} Relative Importance is {}'.format(*rows)) for rows in sorted(input_importance)];
```
